```python
def getword(word):
    b="|"
    for i in range(len(word)):
        b = b+ str(word[i].center(6))+"|"
    return (b)

def print_one_line(v):
    char=len(v)
    sk1= f'{" "*6}|'
    sk2= f'{"_"*6}|'
    print(f'\t|{sk1*char}')
    print(f"\t{getword(v)}")
    print(f'\t|{sk2*char}')
    
def print_all(v):
    char = len(v[0])
    sk1 = f'{"_"*6} '
    print("\n")
    print(f'\t {sk1*(char)}')
    for i in range(6):
        print_one_line(v[i])
    print("\n")


def get_default(length):
    default_list=[]
    for i in range(6):
        default_list.append(list(f'{"*"*length}')) 
    return default_list

def has_only_text(text: str) -> bool:
    return any(c for c in text if text.isalpha())
```

# Rules
#### There is a 5 letter dictionary word generated on random every time
#### You are provided 6 tries to guess the word
#### When you enter any word:
<ol>
<li>any letter matching wil be shown on the correct location in the word, based on your try count</li>
<li>any letter contained in the word but at incorrect location will be shown with an asterix (*)</li>
</ol>

#### If you enter any alphanumeric character or special character or incorrect word length, you are asked to enter the word again, its not counted towards your tries.
#### If you are not able to guess the word in 6 tries, its printed in the end.

```python
## generate a random word
r = RandomWords()
answer = list(r.get_random_word(hasDictionaryDef="true", minLength=5, maxLength=5).lower())

## set some default variables and print the default matrix
victory=False
words = get_default(len(answer))
print_all(words)
print(f"Please Guess a {len(answer)} letter word.")

for tries in range(6):
    ## Keep asking for input untill correct length word entered
    while True:
        x = list(input("enter your word):").lower())
        if((len(x)!=len(answer)) | (not has_only_text(''.join(x)))):
            print(f'lenx = {len(x)} len answer = {len(answer)}')
            print('Sorry you have entered incorrect length word or alphanumeric/special character, please try again')
        else:
            break
    print(f"This was your Try#:{tries+1}, you entered: {x}")
    ## check if the characters in the word match and replace them in the matrix
    ## If you guess the answer the loop break
    for i in range(5):
        if(answer==x):
            print(f'You have found the solution: {answer}')
            victory=True
            words[tries] = answer
            break
        elif(answer[i]==x[i]):
                words[tries][i]=x[i]
        elif((x[i] in answer) & (answer[i]!=x[i])):
                words[tries][i]=x[i]+"*"
    print_all(words)
    if victory:
        break
## If you are not able to guess the answer, its printed at the end
if not victory:
    print('You were not able to get the answers in 6 tries')
    print(f'Solution : {answer}')
```
